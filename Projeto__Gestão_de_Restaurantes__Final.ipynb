## Apresentação do Projeto

O projeto "Sistema de Gestão de Restaurantes" é uma aplicação em Python que permite a gestão de restaurantes em uma plataforma. Com esta ferramenta, os usuários podem adicionar, editar ou remover restaurantes, bem como gerenciar os menus de cada restaurante, incluindo o nome dos alimentos e seus preços. Este projeto foi desenvolvido com o objetivo de fornecer aos alunos um caso de uso real e prático, permitindo-lhes aplicar os conceitos básicos de programação que aprenderam, como variáveis, estruturas condicionais, estruturas de repetição, listas e funções.

## Funcionalidades do Projeto

O sistema oferece as seguintes funcionalidades:

1. **Gestão de Restaurantes:**
   - Adicionar um restaurante com informações como nome, endereço, telefone e tempo de entrega.
   - Editar informações de um restaurante existente.
   - Remover um restaurante da plataforma.

2. **Gestão de Cardápio:**
   - Para cada restaurante, os usuários podem adicionar, editar ou remover itens do cardápio.
   - Cada item do cardápio possui um nome e um preço.

3. **Apresentação de Informações:**
   - Os usuários podem visualizar a lista de todos os restaurantes registrados na plataforma, incluindo seus detalhes.
   - Para cada restaurante, é possível visualizar o cardápio completo, listando os alimentos disponíveis e seus preços.

4. **Funcionalidades Adicionais (Opcionais):**
   - Os alunos são encorajados a aprimorar o projeto com funcionalidades adicionais, como a capacidade de buscar restaurantes por nome ou endereço, calcular o tempo médio de entrega, ou criar uma interface de linha de comando mais amigável.

- **Estruturas de Repetição:** Para facilitar a gestão de múltiplos restaurantes e itens de cardápio, os alunos usarão estruturas de repetição para percorrer listas e realizar ações em lote.

- **Listas:** As informações sobre restaurantes e seus cardápios serão armazenadas em listas, permitindo a manipulação e organização dos dados.

```python
def main():
    lista_de_restaurantes = list()
    condicao = True

    while condicao:
        menu_inicial()
        indice_menu_inicial = input('Escolha a opção que você deseja: ').strip()
        if indice_menu_inicial == "1":  #Ver Restaurantes
            informacoes_de_restaurantes(lista_de_restaurantes)
        elif indice_menu_inicial == "2":    # Adicionar Restaurantes
            lista_de_restaurantes.append(adicionar_restaurante())
        elif indice_menu_inicial == "3":    #Editar Restaurantes
            editar_restaurante(lista_de_restaurantes)
        elif indice_menu_inicial == "4":    #Remover Restaurantes
            remover_restaurante(lista_de_restaurantes)
        elif indice_menu_inicial == "5":    #Cardápios
            listar_opcoes_cardapios(lista_de_restaurantes)
        elif indice_menu_inicial == "6":    #Sair
            print('Finalizando o programa.')
            condicao = False
        else:   #Opção Inválida
            print('Opção Inválida.')
    print(lista_de_restaurantes)


def menu_inicial(): # Menu Inicial
    print(f'''{41*'~'}
G E S T Ã O  D E  R E S T A U R A N T E S

    [1] Ver Restaurantes;
    [2] Adicionar Restaurante;
    [3] Editar Restaurante;
    [4] Remover Restaurante;
    [5] Cardápios;
    [6] Sair
{41*'~'}
    ''')


def informacoes_de_restaurantes(lista_de_restaurantes): # Informações de todos os retaurantes cadastrados
    if len(lista_de_restaurantes) > 0:
        print(f"{'---':^40}")
        print(f"{'Lista de Restaurantes':^40}")
        print(f"{'---':^40}")
        for i, e in enumerate(lista_de_restaurantes):
            print('')
            print(f"""[ {i:^40} ]
    - Nome: {e[0]};
    - Endereço: {e[1]};
    - Telefone: {e[2]};
    - Tempo de entrega: {e[3]} minutos.""")
            print('')
        print(f"{'---':^40}")
    else:
        print(f"{'---':^40}")
        print('Não há restaurante cadastrado no sistema.')
        print(f"{'---':^40}")


def adicionar_restaurante():    # Opção de adicionar restaurante e já acrescentar algum item ao cardápio se o usuário preferir
    print(f"{'---':^40}")
    print('Ok. Vamos adicionar um Restaurante.')
    print('')
    nome = input('Digite o nome do restaurante: ').strip()
    endereco = input(f'Digite o endereço do restaurante {nome}: ').strip()
    telefone = input(f'Digite o telefone do restaurante {nome}: ').strip()
    tempo_de_entrega = input('Digite o tempo de entrega (em minutos): ').strip()

    novo_restaurante = [nome, endereco, telefone, tempo_de_entrega, []]

    condicao = True
    while condicao == True:
        print('')
        opcao = input(f'Quer adicionar um item ao cardápio do Restaurante "{novo_restaurante[0]}" [S/N]: ').strip().upper()[0]
        if opcao == "S":
            novo_item = adicionar_item_cardapio()
            novo_restaurante[4].append(novo_item)
            print(f"{'---':^40}")
        elif opcao == "N":
            print('Entendido.')
            print(f"{'---':^40}")
            condicao = False
        else:
            print('Não entendi. Responda com S ou N.')
            print(f"{'---':^40}")

    return novo_restaurante


def adicionar_item_cardapio():
    nome = input('Digite o nome do item: ').strip()
    preco = input(f'Digite o preço do item "{nome}": ').strip()

    novo_item = [nome, preco]

    return novo_item


def listar_nomes_restaurantes(lista_de_restaurantes):
    print(f"{'---':^40}")
    print(f"{'LISTA DE RESTAURANTES':^40}")
    print(f"{'---':^40}")

    for i, e in enumerate(lista_de_restaurantes):
        print(f"[{i:^3}] - {e[0]}")
        print('')

    print(f"{'---':^40}")


def editar_restaurante(lista_de_restaurantes):  # Menu de edição dos restaurantes cadastrados

    if len(lista_de_restaurantes) > 0:
        listar_nomes_restaurantes(lista_de_restaurantes)
        editar = 'editar'
        while editar.isnumeric() == False:
            editar = input('Escolha o número do restaurante que você quer EDITAR: ')
            print('')
        editar = int(editar)

        if editar not in range(0, len(lista_de_restaurantes)):
                print('Número de restaurante INVÁLIDO!')
                print(f"{'-':-^40}")
        else:
            opcao = input(f'Quer editar o nome {lista_de_restaurantes[editar][0]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[editar][0] = input('Digite o nome do restaurante: ').strip()

            print(f"{'-':-^40}")
            opcao = input(f'Quer editar o endereço {lista_de_restaurantes[editar][1]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[editar][1] = input('Digite o endereço do restaurante: ').strip()

            print(f"{'-':-^40}")
            opcao = input(f'Quer editar o telefone {lista_de_restaurantes[editar][2]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[editar][2] = input('Digite o telefone do restaurante: ').strip()

            print(f"{'-':-^40}")
            opcao = input(f'Quer editar o tempo de entrega {lista_de_restaurantes[editar][3]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[editar][3] = input('Digite o tempo de entrega do restaurante: ').strip()

            print(f"{'-':-^40}")
            print(f'Restaurante {lista_de_restaurantes[editar][0]} editado com SUCESSO.')
            print(f"{'-':-^40}")
    else:
        print(f"{'---':^40}")
        print('Não há restaurante cadastrado no sistema.')
        print(f"{'---':^40}")



def remover_restaurante(lista_de_restaurantes): # Opção de remoção de restaurante cadastrado
    if len(lista_de_restaurantes) == 0:
        print(f"{'---':^40}")
        print('Não há restaurante cadastrado no sistema.')
        print(f"{'---':^40}")
    else:
        listar_nomes_restaurantes(lista_de_restaurantes)
        remover = 'remover'
        while remover.isnumeric() == False:
            remover = input('Escolha o número do restaurante que você quer REMOVER: ')
            print('')
        remover = int(remover)
        opcao = input(f'Tem certeza que quer remover o Restaurante "{lista_de_restaurantes[remover][0]}" [S/N]: ').strip().upper()[0]
        if opcao == 'S':
            nome_restaurante_removido = lista_de_restaurantes[remover][0][:]
            del lista_de_restaurantes[remover]
            print('')
            print(f'O Restaurante "{nome_restaurante_removido}" foi REMOVIDO com SUCESSO.')
            print(f"{'---':^40}")
        elif opcao == 'N':
            print('')
            print(f'Ok. Restaurante "{lista_de_restaurantes[remover][0]}" não foi removido.')
            print(f"{'---':^40}")
        else:
            print('')
            print('Opção INVÁLIDA.')
            print(f"{'---':^40}")



def listar_opcoes_cardapios(lista_de_restaurantes): # Listar as opções de Cardápio do restaurante a ser escolhido
    if len(lista_de_restaurantes) > 0:
        listar_nomes_restaurantes(lista_de_restaurantes)
        cardapio = 'cardapio'
        while cardapio.isnumeric() == False:
            cardapio = input('Escolha o número do restaurante que você quer ver as opções de cardápio: ')
            print('')
        cardapio = int(cardapio)
        if len(lista_de_restaurantes[cardapio][4]) > 0:
            listar_cardápio(lista_de_restaurantes, cardapio)
            cardapio_por_restaurante(lista_de_restaurantes, cardapio)
        else:
            cardapio_por_restaurante(lista_de_restaurantes, cardapio)
    else:
        print(f"{'---':^40}")
        print('Não há restaurante cadastrado no sistema.')
        print(f"{'---':^40}")


def cardapio_por_restaurante(lista_de_restaurantes, cardapio):  # Menu de edição do cardápio do restaurante
    print(f'''Opções de cardápio do Restaurante {lista_de_restaurantes[cardapio][0]}

    [1] Ver cardápio;
    [2] Adicionar item ao cardápio;
    [3] Editar item do cardápio;
    [4] Remover item do cardápio;
    [5] Voltar ao Menu Inicial.
    ''')

    opcao = 'opcao'
    while opcao.isnumeric() == False:
        opcao = input(f'Escolha o número da opção de cardápio do Restaurante "{lista_de_restaurantes[cardapio][0]}": ')
        print('')
    opcao = int(opcao)

    if opcao == 1: # Ver cardápio
        listar_cardápio(lista_de_restaurantes, cardapio)
    elif opcao == 2: # Adicionar item ao cardápio
        condicao = True
        while condicao == True:
            print('')
            opcao = input(f'Quer adicionar um item ao cardápio do Restaurante "{lista_de_restaurantes[cardapio]}" [S/N]: ').strip().upper()[0]
            if opcao == "S":
                novo_item = adicionar_item_cardapio()
                lista_de_restaurantes[cardapio].append(novo_item)
            elif opcao == "N":
                print('Ok. Voltando para o Menu Inicial.')
                condicao = False
            else:
                print('Não entendi. Responda com S ou N.')
    elif opcao == 3: # Editar item ao cardápio
        editar_item_cardapio(lista_de_restaurantes, cardapio)
    elif opcao == 4: # Remover item ao cardápio
        remover_item_cardapio(lista_de_restaurantes, cardapio)
    else: # Voltar ao Menu Inicial
        print('Voltando ao Menu Inicial.')


def editar_item_cardapio(lista_de_restaurantes, cardapio):  # Edição de itens de um cardápio
    if len(lista_de_restaurantes[cardapio]) > 0:
        listar_cardápio(lista_de_restaurantes, cardapio)
        editar = 'editar'
        while editar.isnumeric() == False:
            editar = input(f'Escolha o número do item do Restaurante "{lista_de_restaurantes[cardapio][0]}" que você quer EDITAR: ')
            print('')
        editar = int(editar)
        if editar not in range(0, len(lista_de_restaurantes[cardapio][4])):
                print('Número do item é INVÁLIDO!')
                print(f"{'-':-^40}")
        else:
            opcao = input(f'Quer editar o nome do item {lista_de_restaurantes[cardapio][4][editar][0]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[cardapio][4][editar][0] = input('Digite o nome do item: ').strip()

            print(f"{'-':-^40}")
            opcao = input(f'Quer editar o preço do item {lista_de_restaurantes[cardapio][4][editar][0]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                lista_de_restaurantes[cardapio][4][editar][1] = input(f'Digite o preço do item "{lista_de_restaurantes[cardapio][4][editar][0]}": R$ ').strip()

            print(f"{'-':-^40}")
    else:
        print(f"{'-':-^40}")
        print(f'Não há item do Restaurante "{lista_de_restaurantes[cardapio][0]}" cadastrado no sistema.')
        print(f"{'-':-^40}")


def listar_cardápio(lista_de_restaurantes, cardapio):   #Listar o Cardápio do restaurante escolhido

    if len(lista_de_restaurantes[cardapio][4]) > 0:

        print(f"{'---':^40}")
        print(f"{lista_de_restaurantes[cardapio][0]:^40}")
        print(f"{'---':^40}")
        for i, e in enumerate(lista_de_restaurantes[cardapio][4]):
            print(f"[{i:^3}] Nome: {e[0]:<30} R$ {e[1]}")
            print(f"{'-':-^40}")


def remover_item_cardapio(lista_de_restaurantes, cardapio): # Remover allgum item de cardápio do restaurante escolhido
    if len(lista_de_restaurantes[cardapio][4]) > 0:
        listar_cardápio(lista_de_restaurantes, cardapio)
        remover = 'remover'
        while remover.isnumeric() == False:
            remover = input(f'Escolha o número do item do Restaurante "{lista_de_restaurantes[cardapio][0]}" que você quer REMOVER: ')
            print('')
        remover = int(remover)
        if remover not in range(0, len(lista_de_restaurantes[cardapio][4])):
                print('Número do item é INVÁLIDO!')
                print(f"{'-':-^40}")
        else:
            opcao = input(f'Tem certeza que quer remover o item {lista_de_restaurantes[cardapio][remover][0]} [S/N]: ').strip().upper()[0]
            if opcao == 'S':
                nome_item_removido = lista_de_restaurantes[cardapio][remover][0][:]
                del lista_de_restaurantes[cardapio][remover]
                print('')
                print(f'O Restaurante "{nome_item_removido}" foi REMOVIDO com SUCESSO.')
                print(f"{'---':^40}")

            elif opcao == 'N':
                ('')
                print(f'Ok. O item "{lista_de_restaurantes[cardapio][remover][0]}" não foi removido.')
                print(f"{'---':^40}")



main()
```
